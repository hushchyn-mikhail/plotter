```python
%matplotlib inline
import matplotlib.pyplot as plt

import pandas
import numpy
```

```python
data = "m20_ptmin300_ptmax10000"
name = "la_" + data
```

```python
df = pandas.read_csv('plots/'+name+'.csv')
window = df['X'].values
score = df['ScoreHack2D'].values
reco_eff = df['RecoEff'].values
clone_r = df['CloneRate'].values
ghost_r = df['GhostRate'].values
```

```python
plt.figure(figsize=(12, 8))

plt.plot(window, score, label='ScoreHack2D', linewidth=4, color='r')
plt.plot(window, reco_eff, label='Reco Eff.', linewidth=4, color='gold')
plt.plot(window, clone_r, label='Clone Rate', linewidth=4, color='blue')
plt.plot(window, ghost_r, label='Ghost Rate', linewidth=4, color='green')

plt.title('Method: Linear Approximation', size=20)
plt.xlabel('Method Parameter: window_width', size=20)

plt.xticks(size=20)
plt.yticks(size=20)

plt.ylim(0, 1)

plt.legend(loc='best', prop={'size':20})
plt.grid(b=1)

plt.savefig('plots/'+name+'.png',format='png')
plt.show()
```
